```python
import pandas as pd
```

#  OCORRENCIA

```python
df1=pd.read_excel('Analise Lotes - Copia.xlsx',sheet_name='Ocorrencia')
df1
```

#  RELACOES

```python
df2=pd.read_excel('Analise Lotes - Copia.xlsx',sheet_name='Relacoes')
df2
```

#  ESTOQUE

```python
df3=pd.read_excel('Analise Lotes - Copia.xlsx',sheet_name='Estoque')
df3
```

```python
df_todas_as_planilhas=pd.concat([df1,df2,df3])
df_todas_as_planilhas
```

```python
quadralote=df_todas_as_planilhas.groupby(['Q-L','OCORRENCIA','DATA']).sum().unstack()
quadralote
```

```python
quadralote.to_excel('Result.xlsx')
```
