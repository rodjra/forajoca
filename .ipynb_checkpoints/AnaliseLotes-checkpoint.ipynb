# Observações
- Arquivo 'Esboço.xlsx' dá pra ter uma idéia de como ficar, está no esquema separado por lote e por data, com a Ocorrencia de cada lote na Data do que aconteceu.


```python
import pandas as pd
import numpy as np
```

#  OCORRENCIA

```python
ocorrencia=pd.read_excel('Analise Lotes - Copia.xlsx',sheet_name='Ocorrencia')
ocorrencia=ocorrencia.drop(['QUADRA'], axis=1)
ocorrencia=ocorrencia.drop(['LOTE'], axis=1)
display(ocorrencia)
ocorrencia.info()
```

#  RELACOES

```python
relacoes=pd.read_excel('Analise Lotes - Copia.xlsx',sheet_name='Relacoes')
relacoes=relacoes.drop(['QUADRA'], axis=1)
relacoes=relacoes.drop(['LOTE'], axis=1)
relacoes=relacoes.drop(['OCORRENCIA'], axis=1)
relacoes['OCORRENCIA']=np.nan  # Criar coluna OCORRENCIA
relacoes['OCORRENCIA']=relacoes['OCORRENCIA'].map({np.nan:'VENDIDOS'},na_action=None)  # Preencher NaN com 'VENDIDOS'                  
display(relacoes)
relacoes.info()
```

#  ESTOQUE

```python
estoque=pd.read_excel('Analise Lotes - Copia.xlsx',sheet_name='Estoque')
estoque=estoque.drop(['QUADRA'], axis=1)
estoque=estoque.drop(['LOTE'], axis=1)
display(estoque)
estoque.info()
```

```python
todas=pd.concat([ocorrencia,relacoes,estoque])
display(todas)
todas.info()
```

```python
todas=todas.groupby(['Q-L','DATA']).sum().unstack()
display(todas)
todas.info()
```
